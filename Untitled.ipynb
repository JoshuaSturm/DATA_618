```python
import numpy as np
import pandas as pd
import requests
import os
import urllib
import json
import math
import datetime as dt
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Activation, Dense, Dropout, LSTM
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
###########################
#
# Read in AlphaVantage data
#
###########################

data_source = 'alphavantage'
ticker = "AAL"
api_key = ''
url_string = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=%s&outputsize=full&apikey=%s"%(ticker,api_key) 
file_to_save = 'stock_market_data-%s.csv'%ticker

if not os.path.exists(file_to_save): 
    with urllib.request.urlopen(url_string) as url: 
        data = json.loads(url.read().decode()) 
        data = data['Time Series (Daily)'] 
        df = pd.DataFrame(columns=['Date','Low','High','Close','Open']) 
        for k,v in data.items(): 
            date = dt.datetime.strptime(k, '%Y-%m-%d') 
            data_row = [date.date(),float(v['3. low']),float(v['2. high']), float(v['4. close']),float(v['1. open'])] 
            df.loc[-1,:] = data_row 
            df.index = df.index + 1 
            print('Data saved to : %s'%file_to_save)

df.to_csv(file_to_save)
```

```python
# If file is already saved, load and format it

kaggledf = pd.read_csv(os.path.join('Data','Stocks','hpq.us.txt'),delimiter=',',usecols=['Date','Open','High','Low','Close']).sort_values('Date')
alphadf = pd.read_csv('stock_market_data-AAL.csv').sort_values('Date')

# Check the first few rows
kaggledf.head()
alphadf.head()
```

```python
# Initialize seaborn
sns.set()
plt.figure(figsize = (18,9)) 
plt.plot(range(kaggledf.shape[0]),(kaggledf['Low']+kaggledf['High'])/2.0) 
plt.xticks(range(0,kaggledf.shape[0],500),kaggledf['Date'].loc[::500],rotation=45) 
plt.xlabel('Date',fontsize=18) 
plt.ylabel('Mid Price',fontsize=18) 
plt.show()
```

At this point, I was going to split the data as I normally would - by random sampling a majority to assign for training, and the remainder for testing (sklearn.model_selection.train_test_split makes this very easy.) However, after further reading, I realized that modelling time series data is unique, in that the data is ordered, and must remain sequential. So I will sort the data by the `Date` column, and then take the first 70% of rows to be for training, and the remainder for testing.

```python
# Fix a random seed to ensure consistent results
np.random.seed(94)

# Since we have two values, we'll use the mean of both, i.e. ((low + high) / 2.0)
mean_price = (kaggledf['Low'].values + kaggledf['High'].values) / 2.0

# Split the dataset into two parts - 70% training, 30% testing
tscv = TimeSeriesSplit(n_splits=3)

for train_index, test_index in tscv.split(mean_price):
#     print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = mean_price[train_index], mean_price[test_index]
#     y_train, y_test = y[train_index], y[test_index]

# Reshape the data
train = X_train.reshape(-1, 1)
test = X_test.reshape(-1, 1) 

# Normalize the data so that all values are within [0,1]
scaler = MinMaxScaler(feature_range = (0, 1))

train = scaler.fit_transform(train)
test = scaler.fit_transform(test)
```

We next need to "window" our data. We establish an input size, which will define how large a sample of earlier records we will use to predict an output (whose window size will be defined, too.)
The window is shifted along after each iteration, using new data to predict the next set of outcomes.

```python
def window_data(data, timestep):
    '''
    This function will 'window' our data, and reconfigure our array into a new dataset
    with our defined timestep (window size)
    '''
    dataX = []
    dataY = []
    for i in range(len(data)-timestep):
        dataX.append(data[i:(i + timestep), 0])
        dataY.append(data[i + timestep, 0])
    return np.array(dataX), np.array(dataY)
```

```python
# Define the number of timesteps, aka window size
timestep = 20

trainX, trainY = window_data(train, timestep)
testX, testY = window_data(test, timestep)

# Reshape the data
trainX = np.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
testX = np.reshape(testX, (testX.shape[0], testX.shape[1], 1))
```

```python
print(trainX.shape[0])
print(testX.shape[0])
print(trainY.shape[0])
print(testY.shape[0])
```

```python
# Innitialize the model
model = Sequential()

# Add a hidden layer
model.add(
    LSTM(
        units = 50,
        input_shape = (trainX.shape[1], 1),
        return_sequences = True
    )
)
model.add(
    Dropout(0.2)
)

# Add a second hidden layer
model.add(
    LSTM(
        100,
        return_sequences = False
    )
)
model.add(
    Dropout(0.2)
)

# Add a dense layer
model.add(
    Dense(
        units = 1,
        activation = 'sigmoid'
    )
)
# Loss is calculated using the mean squared error,
# using the adam optimizer,
# and checking the accuracy of each prediction

model.compile(loss="mse", 
              optimizer="adam",
              metrics = ['accuracy'])

output = model.fit(trainX, trainY, batch_size=20, epochs=10)
score = model.evaluate(testX, testY, batch_size=20)
```

```python
plt.plot(output.history['loss'])
plt.plot(output.history['acc'])
```

```python
trainScore = model.evaluate(trainX, trainY, batch_size=16, verbose=0)
print('Train Score: ', trainScore)
testScore = model.evaluate(testX, testY, batch_size=16, verbose=0)
print('Test Score: ', testScore)
```

```python
y_test_inv = scaler.inverse_transform([testY])
mean_price_inv = mean_price.reshape(-1,1)

pred_train = model.predict(trainX)
pred_train = scaler.inverse_transform(pred_train).reshape(-1,1)

pred_test = model.predict([testX]) #pred_test.shape = (num_rows, TIME_AHEAD)
pred_test = scaler.inverse_transform(pred_test).reshape(-1,1)

score = math.sqrt(mean_squared_error(y_test_inv[0], pred_test[:,0]))
print('Test RMSE: %.2f' % (score))
```

```python
plt.figure(figsize=(20,10))
plt.plot(mean_price_inv, color = 'black', label = 'Actual Stock Price')
plt.plot(np.concatenate((pred_train, pred_test)), color = 'green', label = 'Predicted Stock Price')
plt.legend()
plt.show()
```

As can be seen in the plot above, our model starts off quite similar to the actual value, albeit scaled higher. As the model trains new data, and learns from the price history, we can see that its accuracy has greatly improved towards the end, and is nearly aligned with the target graph.
